```python
# 测试Cookie是否生效，如正常因打印出“xx的帐号”字符串
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36'}
url = 'https://www.douban.com'
r = requests.get(url, cookies = cookies, headers = headers).text

soup = BeautifulSoup(r, 'lxml')
pattern = soup.find('span')
for i in pattern:
    print(i.string)
```

```python
import time
import random

attrs = {
    'class': 'short',
}
# 只爬取500条，爬多了会触发豆瓣反爬机制
with open('comment.txt', 'w', encoding='utf-8') as f:
    for num in range(0, 500, 20):
        url = 'https://movie.douban.com/subject/ + str(doubanID) + /comments?start=' + str(num) + '&amp;limit=20&amp;sort=new_score&amp;status=P'
                r = requests.get(url, headers=headers, cookies=cookies).text
                soup = BeautifulSoup(r, 'lxml')
                pattern = soup.find_all(attrs = attrs)
                for i in pattern:
                    print(i.string)
                    f.write(i.string + '\n')
        time.sleep(1 + float(random.randint(1, 100)) / 20)
```

```python
# 测试抓取结果，并导入 dataframe 中
import pandas as pd

content = pd.read_csv('comment.txt', encoding='utf-8', names =["reviews"])

content.head()
```

```python
# jieba分词，去停用词
import jieba

stopwords = []
text = open("doubanreviews.txt", 'w')
with open("stopwords.txt", "r", encoding="utf8") as f:
    for w in f:
        stopwords.append(w.strip())
for i in range(len(content)):        
    words = jieba.cut(content["reviews"][i], cut_all=False, HMM=True)
    for word in words:
        if word not in stopwords:
            text.write(word)
            text.write(" ")
text.close()
```

```python
# 制作词云
from wordcloud import WordCloud
import matplotlib.pyplot as plt
chtext = ''
with open("doubanreviews.txt", 'r') as f:
    for line in f.readlines():
        line = line.strip('\n')
        chtext += line
wordcloud = WordCloud(scale=4, background_color='white',font_path="PingFang.ttc").generate(chtext)
%pylab inline
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
